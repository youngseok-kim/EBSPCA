```julia
using PyPlot
using Roots
using Distributions
using LowRankApprox
include("ash.jl");
```

```julia
function gen_spca(n,p; s = ceil(Int,p/5), signal = 1)
    u = randn(n); v = randn(p); v[s+1:p] = 0;
    u = u/norm(u);
    v = signal * s * v / norm(v);
    X = randn(n,p) + u*v';
    
    return X, u, v
end
```

```julia
srand(2);
X, u_true, v_true = gen_spca(400,200, s = 40, signal = 1);
fig, ax = subplots(1,2, figsize = (10,4))
subplot2grid((10,10), (0,0), rowspan=10, colspan=5)
pcolormesh(X, cmap = "PuBu"); colorbar();
title("heatmap of X"); xlabel("column index (v)"); ylabel("row index (u)")
subplot2grid((10,10), (0,6), rowspan=10, colspan=5)
plot(v_true);
title("sparse principal eigenvector"); xlabel("v_i"); ylabel("value at index i");
```

```julia
function spca(X; iter = 100, init = "by_svd")
    # initialize
    n,p = size(X);
    if init == "by_svd"
        temp = svds(X, nsv = 1)[1];
        u = temp[:U]; u2 = u.^2;
        v = temp[:V].* temp[:S]'; v2 = v.^2;
    elseif init == "by_random"
        u = randn(n); u = u/norm(u); v = X'*u; u2 = u.^2; v2 = v.^2;
    else
        error("Error")
    end
    X2 = X.^2;
    R2 = u2 * v2' + X2 - 2 * (u * v') .* X;
    tau = repmat(1./mean(R2,1),size(R2,1),1);
    
    # loop start
    for i = 1:iter
        
        # update u
        s2 = 1./ (tau * v2);
        x = ((X .* tau) * v) .* s2;
        u = x ./ (s2 + 1);
        u2 = u.^2 + (s2 ./ (s2 + 1));
        
        # update v
        s2 = 1./ (tau' * u2);
        x = ((X .* tau)' * u) .* s2;
        v,v2 = ash(x,s2);
        
        # update tau
        R2 = u2 * v2' + X2 - 2 * (u * v') .* X;
        tau = repmat(1./mean(R2,2),1,size(R2,2));
        
        if rem(i,100) == 0
            print(i); println("-th iteration done");
        end
    end
    
    return u, u2, v, v2, tau
end

function post_median(X, output)
    u,u2,v,v2,tau = output;
    s2 = 1./ (tau' * u2);
    x = ((X .* tau)' * u) .* s2;
    temp = ash(x,s2);
    comp_post_prob = temp[5]; comp_post_mean = temp[6]; comp_post_sd2 = temp[7];
    
    ind = find(comp_post_prob[:,1] .< 0.5);
    post_median = zeros(length(x));
    for i in ind
        a = comp_post_prob[i,2:end];
        b = comp_post_mean[i,2:end];
        c = comp_post_sd2[i,2:end];
        d = comp_post_prob[i,1];
        # post_cdf
        function post_cdf(x)
            return a' * cdf.(Normal(),(x - abs.(b))./sqrt.(c)) + d - 0.5
        end
        post_median[i] = fzero(post_cdf,-1,1e2) * sign(x[i]);
    end
    return post_median
end

function svd_uv(X)
    temp = svds(X, nsv = 1)[1];
    u = temp[:U];
    v = temp[:V] .* temp[:S]';
    return u, v
end

function change_sign(x)
    if norm(x - v_true) < norm(x + v_true)
        return x
    else
        return -x
    end
end
```

```julia
out = spca(X, iter = 500);
v_postmean = change_sign(out[3]);
v_postmed = change_sign(post_median(X, out));
v_svd = change_sign(svd_uv(X)[2]);
```

```julia
fig, ax = subplots(figsize = (12,12))
subplot(221); plot(v_true); title("true vector"); xlabel("index"); ylabel("v_index")
subplot(222); plot(v_svd); title("first singular vector"); xlabel("index"); ylabel("v_index")
subplot(223); plot(v_postmean); title("posterior mean"); xlabel("index"); ylabel("v_index")
subplot(224); plot(v_postmed); title("posterior median"); xlabel("index"); ylabel("v_index");
```
