```julia
x_range = linspace(0,10,100)
include("autoselectmixgrid.jl")
grid = autoselectmixgrid([10.0;10.0], nv = 0);
temp = ["123"];
for i = 1:10
    plot(x_range,pdf.(Truncated(Normal(0,grid[i] + 1), 0, Inf), x_range))
    g = grid[i] + 1;
    if i == 1
        temp = [" = $g"];
    else
        temp = [temp; " = $g"];
    end
end
legend(temp)
```
