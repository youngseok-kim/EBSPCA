```julia
function gen_spca(n,p; k = 1, s = ceil(Int,p/10), d = exp.(-(1:k)/2))
    U = randn(n,k); V = zeros(p,k);
    for j = 1:k
        ind = ((j-1)*s+1):(j*s)
        V[ind,j] = randn(s);
    end
    U = U./sqrt.(sum(U.^2,1));
    X = randn(n,p) + U*Diagonal(d)*V';
    
    return X, U, V, d
end
```

```julia
function post_median(X, output)
    u,u2,v,v2,tau = output;
    s2 = 1./ (tau' * u2);
    x = ((X .* tau)' * u) .* s2;
    temp = ash(x,s2);
    comp_post_prob = temp[5]; comp_post_mean = temp[6]; comp_post_sd2 = temp[7];
    
    ind = find(comp_post_prob[:,1] .< 0.5);
    post_median = zeros(length(x));
    for i in ind
        a = comp_post_prob[i,2:end];
        b = comp_post_mean[i,2:end];
        c = comp_post_sd2[i,2:end];
        d = comp_post_prob[i,1];
        # post_cdf
        function post_cdf(x)
            return a' * cdf.(Normal(),(x - abs.(b))./sqrt.(c)) + d - 0.5
        end
        post_median[i] = fzero(post_cdf,-1,1e2) * sign(x[i]);
    end
    return post_median
end

function svd_uv(X)
    temp = svds(X, nsv = 1)[1];
    u = temp[:U];
    v = temp[:V] .* temp[:S]';
    return u, v
end

function change_sign(x)
    if norm(x - v_true) < norm(x + v_true)
        return x
    else
        return -x
    end
end
```

```julia
function spca(X; iter = 100, init = "by_svd")
    # initialize
    n,p = size(X);
    if init == "by_svd"
        temp = svds(X, nsv = 1)[1];
        u = temp[:U]; u2 = u.^2;
        v = temp[:V].* temp[:S]'; v2 = v.^2;
    elseif init == "by_random"
        u = randn(n); u = u/norm(u); v = X'*u; u2 = u.^2; v2 = v.^2;
    else
        error("Error")
    end
    X2 = X.^2;
    R2 = u2 * v2' + X2 - 2 * (u * v') .* X;
    tau = repmat(1./mean(R2,1),size(R2,1),1);
    tau_old = 0; err = zeros(iter); i = 0;
    
    # loop start
    for i = 1:iter
        
        # update u
        s2 = 1./ (tau * v2);
        x = ((X .* tau) * v) .* s2;
        u = x ./ (s2 + 1);
        u2 = u.^2 + (s2 ./ (s2 + 1));
        
        # update v
        s2 = 1./ (tau' * u2);
        x = ((X .* tau)' * u) .* s2;
        v,v2 = ash2(x,s2);
        
        # update tau
        R2 = u2 * v2' + X2 - 2 * (u * v') .* X;
        tau = repmat(1./mean(R2,1),size(R2,1),1);
        
        if rem(i,Int(iter/5)) == 0
            print(i); println("-th iteration done");
        end
        
        err[i] = vecnorm(R2);
        if err[i] < 1e-10
            break;
        end
        
        tau_old = copy(tau);
    end
    
    return u, u2, v, v2, tau, err[1:i]
end
```

```julia
function spca_m(Y; iter = 10, k = 2)
    
    # initialization
    n,p = size(Y);
    temp = svds(Y, nsv = k)[1];
    c = sign.(temp[:U][1,:])';
    l = SharedArray((temp[:U] .* c) .* temp[:S]' / sqrt(p)); l2 = SharedArray(l.^2);
    f = temp[:V] * sqrt(p) .* c; f2 = f.^2; Y2 = Y.^2;
    R2 = l2 * f2' + Y2 - 2 * (l * f') .* Y;
    tau = repmat(1./mean(R2,1),size(R2,1),1);
    
    # loop start
    for i = 1:iter
        
        # update loading
        s2 = SharedArray(1./ (tau * f2));
        x = SharedArray(((Y .* tau) * f) .* s2);
        @parallel for j = 1:k
            l[:,j],l2[:,j] = ash(x[:,j],s2[:,j]);
        end
        
        # update factor
        s2 = 1./ (tau' * l2);
        x = ((Y .* tau)' * l) .* s2;
        f = x ./ (s2 + 1);
        f2 = f.^2 + (s2 ./ (s2 + 1));
        
        # update tau
        R2 = l2 * f2' + Y2 - 2 * (l * f') .* Y;
        tau = repmat(1./mean(R2,1),size(R2,1),1);
    end
    
    return l,f,tau
end
```
