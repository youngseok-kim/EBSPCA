```julia
function spca(X; iter = 100, init = "by_svd", precision_type = "rowwise_constant")
    # initialize
    n,p = size(X);
    if init == "by_svd"
        temp = svds(X, nsv = 1)[1];
        u = temp[:U] * sqrt(n); u2 = u.^2;
        v = temp[:V] * temp[:S] / sqrt(n); v2 = v.^2;
    elseif init == "by_random"
        u = randn(n); u = u/norm(u); v = X'*u; u2 = u.^2; v2 = v.^2;
    else
        error("Error")
    end
    X2 = X.^2;
    R2 = update_R2(X, X2, u, u2, v, v2);
    tau = update_tau(R2, precision_type)
    
    # loop start
    for i = 1:iter
        
        # update u
        u, u2 = update_u(X, tau, v, v2, precision_type);
        
        # update v
        v, v2 = update_v(X, tau, u, u2, precision_type);
        
        # update tau
        R2 = update_R2(X, X2, u, u2, v, v2)
        tau = update_tau(R2, precision_type)
        
        if rem(i,Int(iter/5)) == 0
            #print(i); println("-th iteration done");
        end
        
        println(vecnorm(tau))
    end
    
    return u, u2, v, v2, tau
end
```
