```julia
function spca(X; iter = 100, init = "by_svd",
              precision_type = "rowwise_constant",
              return_type = "posterior_mean",
              nv = 20, nullprior = 20,
              convtol = 1e-8,
              verbose = true)
    # initialize
    n,p = size(X);
    if init == "by_svd"
        temp = svds(X, nsv = 1)[1];
        u = temp[:U] * sqrt(n); u2 = u.^2;
        v = temp[:V] * temp[:S] / sqrt(n); v2 = v.^2;
    elseif init == "by_random"
        u = randn(n); u = u/norm(u); v = X'*u; u2 = u.^2; v2 = v.^2;
    else
        error("Error")
    end
    X2 = X.^2; i = 0; temp = 0;
    R2 = update_R2(X, X2, u, u2, v, v2);
    tau = update_tau(R2, precision_type);
    
    # loop start
    for i = 1:iter
        
        v_old = v/norm(v);
        
        # update u
        u, u2 = update_u(X, tau, v, v2, precision_type);
        
        # update v
        v, v2, temp = update_v(X, tau, u, u2, precision_type, nv, nullprior);
        
        # update tau
        R2 = update_R2(X, X2, u, u2, v, v2)
        tau = update_tau(R2, precision_type)
        
        a = norm(v); b = norm(v/norm(v) - v_old);
        if b < convtol
            break;
        end
        
        if verbose == true && rem(i,50) == 0
            print(i); println("-th iteration done with norm(v) = $a and error $b");
        end

    end
    
    if return_type == "posterior_median"
        v_postmed = compute_posterior_median(temp)
        return u, u2, v, v2, tau, i, temp, v_postmed
    end
    
    return u, u2, v, v2, tau, i, temp
end
```

```julia
function compute_posterior_median(temp)
    comp_post_prob = temp[5]; comp_post_mean = temp[6]; comp_post_sd2 = temp[7];
    
    ind = find(comp_post_prob[:,1] .< 0.5);
    post_median = zeros(length(x));
    for i in ind
        a = comp_post_prob[i,2:end];
        b = comp_post_mean[i,2:end];
        c = comp_post_sd2[i,2:end];
        d = comp_post_prob[i,1];
        # post_cdf
        function post_cdf(x)
            return a' * erfc.((x - abs.(b))./sqrt.(2*c)) + d - 0.5
        end
        post_median[i] = fzero(post_cdf,-1,1e2) * sign(temp[1][i]);
    end
    return post_median
end
```

```julia
function spca_m(Y; iter = 10, k = 2)
    
    # initialization
    n,p = size(Y);
    temp = svds(Y, nsv = k)[1];
    c = sign.(temp[:U][1,:])';
    l = SharedArray((temp[:U] .* c) .* temp[:S]' / sqrt(p)); l2 = SharedArray(l.^2);
    f = temp[:V] * sqrt(p) .* c; f2 = f.^2; Y2 = Y.^2;
    R2 = l2 * f2' + Y2 - 2 * (l * f') .* Y;
    tau = repmat(1./mean(R2,1),size(R2,1),1);
    
    # loop start
    for i = 1:iter
        
        # update loading
        s2 = SharedArray(1./ (tau * f2));
        x = SharedArray(((Y .* tau) * f) .* s2);
        @parallel for j = 1:k
            l[:,j],l2[:,j] = ash(x[:,j],s2[:,j]);
        end
        
        # update factor
        s2 = 1./ (tau' * l2);
        x = ((Y .* tau)' * l) .* s2;
        f = x ./ (s2 + 1);
        f2 = f.^2 + (s2 ./ (s2 + 1));
        
        # update tau
        R2 = l2 * f2' + Y2 - 2 * (l * f') .* Y;
        tau = repmat(1./mean(R2,1),size(R2,1),1);
    end
    
    return l,f,tau
end
```
