```julia
function soft_thresh(x, s)
    return max.(abs.(x) - s, 0) .* sign.(x)
end

function normaliz(z)
    z[1:3] = z[1:3]/norm(z[1:3]);
    z[4:6] = z[4:6]/norm(z[4:6]);
    return z
end
```

```julia
C = speye(6) * 48 + [zeros(3,3) -S;-S zeros(3,3)];
a = chol(full(C));
z = Variable(6);
cost = norm(a*z);
solve!(minimize(cost,[norm(z[1:3]) <= 1, norm(z[4:6]) <= 1]), MosekSolver())
z.value
```

```julia
u = randn(1100); t = 1e-4;
D = [A zeros(size(A)); zeros(size(B)) B]
for i = 1:10000
    u = soft_thresh(u - t * D*C*(D'*u), t);
    u = D * normaliz(D'*u); 
end
z = D'*u
```

```julia
alpha = ones(3)/sqrt(3); beta = ones(3)/sqrt(3);
lambda1 = 1; lambda2 = 1; t = 1;
for i = 1:1000
    alpha = soft_thresh(alpha + t * A * (s .* beta), lambda1);
    beta = soft_thresh(beta + t * B * (s .* alpha), lambda2);
end
```
